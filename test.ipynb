```python
# Testando

# Importando as bibliotecas necessárias
from pandas import read_csv, concat
from sklearn.model_selection import train_test_split

# Lendo os dataframes de teste e treino
df_train = read_csv(r'data/california_housing_train.csv')
df_test = read_csv(r'data/california_housing_test.csv')

# Concatenando os dataframes de teste e treino
df = concat([df_test,df_train])

# Separando os dados de treino e teste
x = df['households']
y = df['total_bedrooms']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Instanciando o modelo
model = SimpleLinearRegression()

# Ajustando o modelo
model.fit(x_train, y_train)

# Estimando os valores
class_predict = model.predict(x_train)

# Calculando as métricas
r2_score = model.r2_score(class_predict, y_train)
square_error = model.square_error(class_predict, y_train)
mean_square_error = model.mean_square_error(class_predict, y_train)
model_equation = model.show_equation(False)

sizes = [len(str(r2_score)), len(str(square_error)), len(str(mean_square_error)), len(model_equation)]
max_size = max(sizes)

def median_size():
    median = max_size/2
    return int(median) + 2

print(f'{median_size()*"="} METRICS {median_size()*"="}')
print(f'| R2 Score | {r2_score} ')
print(f'| Square Error | {square_error} ')
print(f'| Mean Square Error | {mean_square_error} ')
print(f'| Model Equation | {model_equation} ')

# Plotando o gráfico
model.regression_plot()
```
